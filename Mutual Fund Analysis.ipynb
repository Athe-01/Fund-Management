```python
import pandas as pd
import openpyxl
import os
import matplotlib.pyplot as plt
import seaborn as sns
from skimpy import skim

def load_data_as_dict(file_path):
    try:
        if not file_path.endswith('.xlsx'):
            raise ValueError("Unsupported file format.")
        workbook = openpyxl.load_workbook(file_path, read_only=False)
        sheet = workbook.active
        data = [list(row) for row in sheet.iter_rows(values_only=True) if any(cell is not None for cell in row)]
        return data
    except Exception as e:
        print(f"Error loading data from {file_path}: {e}")
        return None

def save_tables(tables, base_name):
    try:
        for idx, table in enumerate(tables):
            df = pd.DataFrame(table)
            output_path = f"{base_name}_Table_{idx + 1}.xlsx"
            df.to_excel(output_path, index=False)
            print(f"Table {idx + 1} saved to {output_path}")
    except Exception as e:
        print(f"Error saving tables: {e}")

def create_tables_from_data(data):
    if not data:
        print("No data to process.")
        return []
    tables, current_table, current_length = [], [], len(data[0])
    for row in data:
        row_length = len(row)
        if row_length == current_length:
            current_table.append(row)
        else:
            if current_table:
                tables.append(current_table)
            current_table, current_length = [row], row_length
    if current_table:
        tables.append(current_table)
    return tables

def analyze_missing_values(data, title):
    try:
        missing_values = data.isnull().sum()
        missing_percentage = (missing_values / len(data)) * 100
        missing_df = pd.DataFrame({
            "Column": missing_values.index,
            "Missing Values": missing_values.values,
            "Percentage (%)": missing_percentage.values
        }).sort_values(by="Percentage (%)", ascending=False)
        print(f"Missing Value Analysis for {title}:")
        print(missing_df)
        plt.figure(figsize=(10, 6))
        sns.heatmap(data.isnull(), cbar=False, cmap="binary", yticklabels=False)
        plt.title(f"Missing Values Heatmap for {title}")
        plt.show()
        return missing_df
    except Exception as e:
        print(f"Error analyzing missing values for {title}: {e}")
        return None

def visualize_changes(allocation_changes):
    try:
        total_changes = pd.DataFrame()
        for change in allocation_changes:
            total_changes = pd.concat([total_changes, change], ignore_index=True)
        if total_changes.empty:
            print("No data available for visualization.")
            return
        total_summary = total_changes.groupby("Name of the Instrument").agg(
            Total_Change=("Allocation_Change", "sum"),
            Total_Market_Value=("Market value\n(Rs. in Lakhs)_Curr", "sum"),
            Total_Quantity=("Quantity_Curr", "sum")
        ).reset_index().sort_values(by="Total_Market_Value", ascending=False)
        print(total_summary)
        plt.figure(figsize=(12, 6))
        plt.bar(total_summary["Name of the Instrument"], total_summary["Total_Market_Value"], color='blue')
        plt.xticks(rotation=45)
        plt.title("Total Market Value of Selected Funds")
        plt.xlabel("Funds")
        plt.ylabel("Total Market Value (Rs. in Lakhs)")
        plt.tight_layout()
        plt.show()
        output_file = "total_summary.xlsx"
        total_summary.to_excel(output_file, index=False)
        print(f"Summary saved as '{output_file}'.")
        display_excel_file(output_file)
        print("\nSkimpy Summary:")
        print(skim(total_summary))
    except Exception as e:
        print(f"Error visualizing changes: {e}")

def display_excel_file(file_path):
    try:
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"The file '{file_path}' does not exist.")
        data = pd.read_excel(file_path)
        print(f"\nContents of '{file_path}':")
        print(data)
    except Exception as e:
        print(f"Error displaying Excel file '{file_path}': {e}")

def extract_and_clean_table(file_path, sheet_name):
    try:
        data = pd.read_excel(file_path, sheet_name=sheet_name, header=None, dtype=str)
        header_indices = [i for i, row in data.iterrows() if "Name of the Instrument" in row.values]
        if not header_indices:
            raise ValueError("No header row found.")
        tables = []
        for idx in range(len(header_indices)):
            start = header_indices[idx]
            end = header_indices[idx + 1] if idx + 1 < len(header_indices) else len(data)
            table = data.iloc[start:end].reset_index(drop=True)
            headers = table.iloc[0].fillna("Unknown").tolist()
            table = table.iloc[1:].reset_index(drop=True)
            table.columns = headers
            table = table.dropna(axis=1, how='all')
            tables.append(table)
        return tables
    except Exception as e:
        print(f"Error extracting and cleaning table from {file_path}: {e}")
        return []

def process_mutual_fund_data(file_paths, sheet_name):
    all_tables = []
    for file_path in file_paths:
        tables = extract_and_clean_table(file_path, sheet_name)
        all_tables.extend(tables)
    return all_tables

def compare_allocations(data_frames, fund_names=None):
    all_changes = []
    for i in range(len(data_frames) - 1):
        prev_data = data_frames[i]
        curr_data = data_frames[i + 1]
        if fund_names:
            pattern = '|'.join([name.strip() for name in fund_names])
            prev_data = prev_data[prev_data["Name of the Instrument"].str.contains(pattern, na=False)]
            curr_data = curr_data[curr_data["Name of the Instrument"].str.contains(pattern, na=False)]
        common_columns = list(set(prev_data.columns) & set(curr_data.columns))
        prev_data = prev_data[common_columns]
        curr_data = curr_data[common_columns]
        if "Market value\n(Rs. in Lakhs)" not in common_columns:
            print(f"Skipping comparison for month {i + 1}: Required column 'Market value\n(Rs. in Lakhs)' not found.")
            continue
        monthly_changes = pd.merge(prev_data, curr_data, on="Name of the Instrument", suffixes=("_Prev", "_Curr"), how="outer")
        monthly_changes["Allocation_Change"] = (
            monthly_changes["Market value\n(Rs. in Lakhs)_Curr"].astype(float) -
            monthly_changes["Market value\n(Rs. in Lakhs)_Prev"].astype(float)
        )
        all_changes.append(monthly_changes)
    return all_changes

def main():
    september_path = "/home/oem/Downloads/ZN250 - Monthly Portfolio September 2024.xlsx"
    november_path = "/home/oem/Downloads/ZN250 - Monthly Portfolio November 2024.xlsx"
    september_data = load_data_as_dict(september_path)
    november_data = load_data_as_dict(november_path)
    september_tables = create_tables_from_data(september_data)
    november_tables = create_tables_from_data(november_data)
    if september_tables:
        save_tables(september_tables, "September_Data")
    if november_tables:
        save_tables(november_tables, "November_Data")
    file_paths = [
        "/home/oem/September_Data_Table_1.xlsx",
        "/home/oem/November_Data_Table_1.xlsx",
    ]
    sheet_name = "Sheet1"
    data_frames = process_mutual_fund_data(file_paths, sheet_name)
    if not data_frames:
        print("No data available for processing.")
        return
    for i, df in enumerate(data_frames):
        analyze_missing_values(df, f"Month {i + 1} Data")
    print("Enter the mutual fund names (comma-separated) you want to track:")
    fund_names = input().split(',')
    allocation_changes = compare_allocations(data_frames, fund_names=fund_names)
    print("Allocation Changes:")
    for i, change in enumerate(allocation_changes):
        print(f"\nChanges for month {i + 1}:")
        print(change)
    visualize_changes(allocation_changes)

if __name__ == "__main__":
    main()
```

```python
import os
import pinecone
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
import requests

PINECONE_API_KEY = "Insert the api key"
PINECONE_ENVIRONMENT = "us-east-1"
INDEX_NAME = "mutual-funds-index"
DOCUMENT_DIR = "/home/oem/Downloads/"
EMBEDDING_MODEL = SentenceTransformer("all-MiniLM-L6-v2")
GEMINI_API_KEY = "Insert Key"

def load_documents(directory):
    documents = []
    filenames = []
    metadata = []
    files = [file for file in os.listdir(directory) if file.endswith(('.txt', '.xlsx', '.csv'))]
    for file in tqdm(files, desc="Loading documents"):
        file_path = os.path.join(directory, file)
        filenames.append(file)
        metadata.append({"filename": file})
        if file.endswith(".txt"):
            with open(file_path, "r") as f:
                documents.append(f.read())
        elif file.endswith((".xlsx", ".csv")):
            try:
                import pandas as pd
                df = pd.read_excel(file_path) if file.endswith(".xlsx") else pd.read_csv(file_path)
                documents.append(df.to_string(index=False, header=True))
                metadata[-1]["columns"] = df.columns.tolist()
            except Exception:
                documents.append(f"Could not parse {file}")
    return documents, filenames, metadata

def create_embeddings(documents, model):
    embeddings = []
    for doc in tqdm(documents, desc="Generating embeddings"):
        embeddings.append(model.encode(doc))
    return embeddings

def initialize_pinecone():
    pc = pinecone.Pinecone(api_key=PINECONE_API_KEY, environment=PINECONE_ENVIRONMENT)
    if INDEX_NAME not in pc.list_indexes().names():
        pc.create_index(name=INDEX_NAME, dimension=384, metric="cosine")
    return pc.Index(INDEX_NAME)

def gemini_generate(prompt, api_key):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key={api_key}"
    headers = {"Content-Type": "application/json"}
    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": prompt
                    }
                ]
            }
        ]
    }
    try:
        response = requests.post(url, headers=headers, json=data)
        if response.status_code == 200:
            return response.json().get("candidates", [{}])[0].get("content", {}).get("parts", [{}])[0].get("text", "No response from Gemini.")
        else:
            print(f"Response Content: {response.text}")
            return f"Gemini API Error: {response.status_code}"
    except Exception as e:
        return f"Exception during Gemini API call: {e}"

def generate_response(query, corpus, pinecone_index, embedding_model, top_k=3):
    query_embedding = embedding_model.encode(query).tolist()
    results = pinecone_index.query(vector=query_embedding, top_k=top_k, include_metadata=True)
    retrieved_docs = [corpus[int(match.id)] for match in results.matches]
    context = " ".join(retrieved_docs[:top_k])
    prompt = f"Context: {context}\nQuestion: {query}\nAnswer:"
    return gemini_generate(prompt, GEMINI_API_KEY)

def chat():
    print("Chat with your RAG-powered system! Type 'exit' to quit.")
    documents, filenames, metadata = load_documents(DOCUMENT_DIR)
    embeddings = create_embeddings(documents, EMBEDDING_MODEL)
    pinecone_index = initialize_pinecone()
    for i, embedding in tqdm(enumerate(embeddings), total=len(embeddings), desc="Uploading to Pinecone"):
        pinecone_index.upsert([(str(i), embedding.tolist(), metadata[i])])
    while True:
        query = input("You: ").strip()
        if query.lower() == "exit":
            print("Goodbye!")
            break
        response = generate_response(query, documents, pinecone_index, EMBEDDING_MODEL)
        print(f"RAG: {response}")

if __name__ == "__main__":
    chat()
```
